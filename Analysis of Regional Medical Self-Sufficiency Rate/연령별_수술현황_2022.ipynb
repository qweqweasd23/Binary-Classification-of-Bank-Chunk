```python
# 기본 함수화
def clean_data(x):
    # 필요없는 정보와 결측치 열 제거
    x = x.iloc[2:,:].copy()
    x = x.iloc[:-3,:].copy()
    # 남자 여자 필요없이 총합만 추출
    x.drop((x[(x['연령별 성별'] == '남자') | (x['연령별 성별'] == '여자')]).index, inplace=True)
    # 결측치 열 제거
    x.drop(4, inplace=True)
    #필요없는 열 제거
    x.drop(['Unnamed: 1'], axis=1, inplace=True)
    # 인덱스 리셋
    x.reset_index(drop=True, inplace=True)
    return x
```

```python
# 2차 정제 함수
def df_clean(x):
    # 데이터의 크기에 따라
    if x.shape == (68, 20):
        # 1차 정제후
        x = clean_data(x)
        # 필요한 정보를 넣어주고
        x.iloc[0,7] = x.iloc[0,9]
        # 필요한 정보만 추출
        x = x.iloc[:,[0,1,7,13]].copy()
        return x
    else:
        # 이하 동문
        x = clean_data(x)
        x = x.iloc[:, [0,1,7,13,19]].copy()
        return x
```
